```python
class Node:
    def __init__(self, value):
        self.next = None
        self.prev = None
        self.val = value

class DoubleLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    # Add Element at the Beginning.....
    def insert_beginning(self, val):
        node = Node(val)
        temp = self.head
        temp.prev = node
        node.next = temp
        self.head = node
        self.size += 1

    #Add Element at the End.....
    def insert_end(self, val):
        node = Node(val)
        temp = self.head
        while  temp.next is not None:
            temp = temp.next
        temp.next = node
        node.prev = temp
        self.size += 1

    #Element add
    def add(self, val):
        node = Node(val)
        if self.tail is None:
            self.head = node
            self.tail = node
            self.size += 1

        else:
            self.tail.next = node
            node.prev = self.tail
            self.tail = node
            self.size += 1


    def __remove_node(self,node):
        if node.prev is None:
            self.head = node.next
        else:
            node.prev.next = node.next

        if node.next is None:
            self.tail = node.prev
        else:
           node.next.prev = node.prev
        self.size -= 1

    #Remove Element at the LinkedList...
    def remove(self, value):
        node = self.head
        while node is not None:
            if node.val == value:
                self.__remove_node(node)
                #break
            node = node.next

    #Remove Element at the First Position.....
    def remove_first(self):
        if self.head is not None:
            self.__remove_node(self.head)
            self.size -= 1

    #Remove Element at the End Position....
    def remove_last(self):
        if self.tail is not None:
            self.__remove_node(self.tail)
            self.size -= 1

    # def reverse(self):
    #     tmp = None
    #     cur = self.head
    #     while cur:
    #         tmp = cur.prev
    #         cur.prev = cur.next
    #         cur.next = tmp
    #         cur = cur.prev
    #
    #     if tmp:
    #         self.head = tmp.prev

    #Reverse LinkedList......
    def reverse(self):
        temp = None
        current = self.head
        while current:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev

        if temp:
            self.head = temp.prev

    #Print LinekedList.........
    def __str__(self):
        vals = []
        node = self.head
        while node is not None:
            vals.append(node.val)
            node = node.next
        return f"[{'-> '.join(str(val) for val in vals)}]"

my_list = DoubleLinkedList()
my_list.add(1)
my_list.add(3)
my_list.add(5)
my_list.add(7)
my_list.add(9)
my_list.add(11)
my_list.add(13)
my_list.add(15)
print(my_list)
my_list.remove(11)
print(my_list)
my_list.remove_last()
print(my_list)
print(my_list.size)

my_list.remove_first()
print(my_list)
my_list.reverse()
print(my_list)

my_list.insert_beginning(55)
my_list.insert_beginning(74)
print(my_list)
my_list.insert_end(66)
my_list.insert_end(99)
print(my_list)
print(my_list.size)
```
